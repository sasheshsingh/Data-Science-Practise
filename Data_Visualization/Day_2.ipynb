# Normal Distribution

```r
library(dslabs)
```

```r
data(heights)
```

```r
x <- heights$sex == "Male"
```

```r
x <- heights$height[x]
```

```r
hist(x)
```

```r
average <- mean(x)
```

```r
sd_values <- sd(x)
```

```r
c(average = average, SD = sd_values)
```

```r
plot(density(x))
```

## Standard Unit

```r
z <- (x - mean(x))/sd(x)
```

```r
mean(abs(z)<=1)
```

```r
mean(abs(z)<=2)
```

```r
mean(abs(z)<=3)
```

# Discretization

```r
index <- heights$sex == "Male"
```

```r
x <- heights$height[index]
```

```r
pnorm(79, mean(x), sd(x))
```

```r
plot(density(x))
```

```r
plot(prop.table(table(x)), xlab = "a = Height in inches", ylab = "Pr(x = a)")
```

```r
mean(x <= 68.5) - mean(x <= 67.5)
```

```r
pnorm_f <- function(d){
    pnorm(d,mean(x),sd(x))
}
```

```r
pnorm_f(68.5) - pnorm_f(67.5)
```

```r
mean(x <= 69.5) - mean(x <= 68.5)
```

```r
pnorm_f(69.5) - pnorm_f(68.5)
```

```r
mean(x <= 70.5) - mean(x <= 69.5)
```

```r
pnorm_f(70.5) - pnorm_f(69.5)
```

```r
mean(x <= 70.1) - mean(x <= 69.9)
```

```r
pnorm_f(70.1) - pnorm_f(69.9)
```

```r
mean(x <= 70) - mean(x <= 69)
```

```r
pnorm_f(70) - pnorm_f(69)
```
